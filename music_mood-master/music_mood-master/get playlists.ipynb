#### Get MP3s for each song id passed as a list

```python
import os
import urllib
import numpy as np

def download_mp3s(song_id_list: list, batch_size: int, mood: str):
    if not os.path.exists('mp3s'):
        os.makedirs('mp3s')
    if not os.path.exists('data'):
        os.makedirs('data/mp3s')
    batches = int(np.ceil(len(song_id_list)/batch_size))
    print(f"Num Batches: {batches}")
    for i in range(0, batches):
        print(f"Downloading Batch {i}")
        track_list = song_id_list[(i*batch_size):(i+1)*batch_size]
        tracks = sp.tracks(track_list)
        success = 0
        for track_num in tracks['tracks']:
            url = track_num['preview_url']
            track_id = track_num["id"]
            try:
                urllib.request.urlretrieve(url, f"data/mp3s/{mood}/{track_id}.mp3")
                success += 1
            except Exception:
                pass
        num_tracks = str(len(tracks['tracks']))
        print(f"Successfuly downloaded {success} songs out of {num_tracks} total. Batch Complete.")
            
```

#### Iterate through CSV Files, creating lists of song ids and batch downloading the mp3 files

```python
import glob

for mood in ['calm' , 'energetic', 'happy', 'sad']:
    print("\n" + mood.capitalize())
    path = f"data/{mood}/*.csv"
    for fname in glob.glob(path):
        print(fname)
        data = pd.read_csv(fname)
        songs = np.unique(data.id.tolist())
        print(len(songs))
        download_mp3s(songs, batch_size=20, mood=mood)
#         break
        
```
